```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import math as m  
from sympy import *  
from sympy import lambdify
```

```python
x=Symbol('x')

F=0.2*x**2 +4        #a differentiable function
f_p=F.diff(x)        #first derivative
F=lambdify(x, F)
f_p=lambdify(x, f_p)

x_0=4               
```

```python
def gradient(f, f_prim, x, learning_rate, precision, max_iter):
    iter_count=0
    prv_step=1
    xtp=()
    ytp=()
    while (prv_step > precision ) and (iter_count < max_iter):
        prv_x=x
        x=x-(learning_rate * f_prim(prv_x))
        prv_step=abs(prv_x - x)
        iter_count+=1
        xtp+=(x, )
        ytp+=(f(x), )
        print(iter_count,')', x)
        
    print('the local min: ', x)
    
    xp = np.linspace(-5, 5, num=50)
    fig, ax = plt.subplots(figsize=(5, 2.7))
    ax.plot(xp, (0.2*xp**2 +4), label='cubic')
    plt.scatter(xtp, ytp)
    
```

```python
gradient(F, f_p, x_0, 0.1, 0.00001, 10000)
```

```python
gradient(F, f_p, x_0, 0.2, 0.00001, 10000)
```
